```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
plt.figure()


origin_img = cv.imread("img/lena_gray_512.tif", cv.IMREAD_GRAYSCALE)
print(origin_img.shape)
origin_img_doub = origin_img.astype(np.float32)
mask = np.zeros_like(origin_img_doub)
plt.subplot(1, 3, 1)
plt.imshow(origin_img, cmap="gray", interpolation=None)


# 以255，255为中心半径为150的圆，其值全部置为1
# 遍历mask
for i in range(0, 512):
    for j in range(0, 512):
        if (i - 255) * (i - 255) + (j - 255) * (j - 255) < 150* 150:
            mask[i, j] = 1

res = origin_img*mask
plt.subplot(1, 3, 2)
plt.imshow(res, cmap="gray", interpolation=None)


res = origin_img * np.zeros_like(origin_img_doub)
plt.subplot(1, 3, 3)
plt.imshow(res, cmap="gray", interpolation=None)

plt.show()
```
