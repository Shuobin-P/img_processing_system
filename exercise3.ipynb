```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
plt.figure()


origin_img = cv.imread("img/lena_gray_512.tif", cv.IMREAD_GRAYSCALE)
print(origin_img.shape)
print(origin_img)
```

```python
# 512*512
all_num = np.linspace(1, 0, num=512)
mask = np.tile(all_num[:, np.newaxis], (1, origin_img.shape[1]))
print(mask)
res = mask * origin_img
plt.figure()
plt.imshow(res, cmap="gray", interpolation=None)
plt.show()
```
