```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt


def extract(image: np.ndarray, x_start, x_end, y_start, y_end)-> np.ndarray:
    # 把指定范围的像素点提取出来
    return image[x_start: x_end, y_start: y_end]    
    

origin_img = cv.imread("img/cameraman.tif")
res_matrix = extract(origin_img, 70, 160, 190, 290)
plt.figure()
plt.imshow(res_matrix, cmap='gray',interpolation='none')
plt.show()
```
